# Importando Dataframe

```python
uri = "/content/ocorrencia.csv"
```

```python
valores_ausentes = ['**','###!','####','****','*****','NULL']
```

```python
df = pd.read_csv(uri, sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes, engine='python', encoding='ISO-8859-1')
df.head(10)
```

```python
df.dtypes
```

Como temos muitas colunas, vou excluir algumas para facilitar a análise

```python
df.drop(columns = ['codigo_ocorrencia1', 
                   'codigo_ocorrencia3', 
                   'codigo_ocorrencia4', 
                   'ocorrencia_latitude', 
                   'ocorrencia_longitude', 
                   'ocorrencia_pais', 
                   'ocorrencia_pais', 
                   'investigacao_aeronave_liberada', 
                   'investigacao_status', 
                   'divulgacao_relatorio_numero', 
                   'divulgacao_relatorio_publicado', 
                   'divulgacao_dia_publicacao', 
                   'total_aeronaves_envolvidas', 
                   'ocorrencia_saida_pista'], inplace = True)
```

```python
schema = pa.DataFrameSchema(
    columns = {
        "codigo_ocorrencia": pa.Column(pa.Int),
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String),
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
        "ocorrencia_aerodromo": pa.Column(pa.String),
        "ocorrencia_dia": pa.Column(pa.DateTime),
        "ocorrencia_hora": pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=False),
        "total_recomendacoes": pa.Column(pa.Int)   
    }
)        
```

```python
df.isna().sum()
```

```python
nan_values = df.loc['codigo_ocorrencia':'total_recomendacoes'].isna()
print(nan_values)
```

```python
df = df.dropna()
```

```python
# df['ocorrencia_uf'] = df['ocorrencia_uf'].replace('***', '--')
df.replace(['**', '***', '###!','####','****','*****','NULL'], pd.NA, inplace=True)
```
